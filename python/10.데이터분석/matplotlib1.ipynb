```python
#  
import numpy as np
x =  np.arange(3)
y = [100, 400, 900]

plt.bar(x,y, color=['r','g','b'], width=1)
plt.plot(x,y)
plt.xticks(x, [2018, 2019, 2020])
#plt.yticks(y, [a,b,c])
plt.show
```

```python
#  
y =  np.arange(3)
x = [100, 400, 900]

plt.barh(y,x, color=['r','g','b'], height=0.7)
plt.plot(x,y)
plt.yticks(y, ['2018', '2019', '2020'])
#plt.yticks(y, [a,b,c])
plt.show
```

```python
n = 50
x = np.random.rand(n)
y = np.random.rand(n)

area = (30*np.random.rand(n))**2
colors = np.random.rand(n)

plt.scatter(x,y,area, colors, cmap='Spectral', alpha=0.5) #cmap : color map
```

```python
x = [0,1,2]
y = [0,1,2]
area = (1000,1000,1000)
plt.plot(x,y)

plt.scatter(x,y,area, alpha=0.5)
```

```python
# 
ratio = [24, 28, 12, 36]
labels = ['a', 'b', 'c', 'd']
wedgeprops = {'width': 0.7, 'edgecolor' : 'k', 'linewidth' : 1}

plt.pie(ratio, labels=labels, wedgeprops = wedgeprops, autopct='%.1f%%', explode=[0,0.1,0,0.15])
plt.show()
```
