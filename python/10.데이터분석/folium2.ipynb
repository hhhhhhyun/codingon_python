```python
import folium
import pandas as pd
import os #파일 주소 확인
print(os.getcwd())
```

```python
data = pd.read_csv("택시승차대 현황.csv", encoding='cp949')
data
```

```python
taxi = data.loc[:, ["위치명", "위도", "경도"]]
taxi
```

```python
m = folium.Map(location=[37.547631, 126.942463], zoom_start=100)

for index, row in taxi.iterrows():
    location = [row["위도"], row["경도"]]
    folium.Marker(location=location, popup=row["위치명"], tooltip=row["위치명"],icon=folium.Icon(color='black', icon='fa-solid fa-taxi', prefix=('fa-solid'))).add_to(m)

m
```
