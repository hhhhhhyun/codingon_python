```python
#실습1. 아래와 같은 그래프를 그려보세요.
import matplotlib.pyplot as plt
```

```python
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
sales_2019, = plt.plot([100,120,140,110,130,150,160,170,180,200,190,210], label = "2019")
sales_2020, = plt.plot([90,110,130,120,140,160,170,160,150,180,200,190], label = "2020" )
plt.legend(ncol=1)
plt.title('Monthly Sales Comparison (2019-2020)')
plt.xlabel('Month', loc='center')
plt.ylabel('Sales', loc='center')
plt.show()
```

```python
#실습2. 아래와 같은 그래프를 그려보세요.
import numpy as np
x =  np.arange(5)
data = [20, 35, 15, 27, 45]
categories = ['Category 1','Category 2','Category 3','Category 4','Category 5']

plt.bar(x,data, width=0.7)
plt.title('Bar Chart')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.xticks(x, categories, rotation=45)
plt.grid(True, axis = 'both')
plt.ylim(0,50) # y축 리미트 걸어서 50까지 나오게해주기기
plt.show
```
