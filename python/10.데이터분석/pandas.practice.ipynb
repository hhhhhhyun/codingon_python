```python
#파이스탁 예제
#chapter05.연산자
#1. low와 high 차이가 100 이상인 날의 고가를 출력하세요.
import pandas as pd
low = pd.Series([10, 200, 200, 400, 600])
high = pd.Series([100, 300, 400, 500, 600])

cond = (high - low) >= 100
high[cond]
```
