```python
import folium
```

```python
m = folium.Map(location = [37.547631, 126.942463], zoom_start=50) #zoom_start : 확대
m
```

```python
m = folium.Map(location = [37.547631, 126.942463], zoom_start=50, tiles = 'cartodbpositron') #tiles: 타일 색을 바꿀 수 있다.     
m
```

```python
m = folium.Map(location = [37.547631, 126.942463], zoom_start=50, width = 800, height=600)
m
```

```python
folium.Marker(location = [37.547631, 126.942463], popup="팝업", tooltip="툴팁").add_to(m)
m
```

```python
# 더 꾸미기
# https://fontawesome.com/v6/icons?o=r&s=solid
folium.Marker(location = [37.547631, 126.942463], popup="팝업", tooltip="툴팁",
              icon=folium.Icon(color='black', icon='fa-solid fa-heart', prefix=('fa-solid'))).add_to(m)
m
```

```python
m.add_child(folium.ClickForMarker(popup="클릭"))
m
```

```python
#m.add_child(folium.ClickForMarker(popup="클릭"))
m.add_child(folium.LatLngPopup())
m
```

```python
folium.CircleMarker([37.547631, 126.942463],
                    radius=100,
                    color='#ff0000', # 초록, #ffffff :하양, #000000: 검정
                    fill_color='#00ff00',
                    popup='CircleMarker' #유저 시점으로 크기 유지(시점에 따라 달라진다.)
                    ).add_to(m)
folium.Circle([37.547631, 126.942463],
                    radius=100,
                    color='#0000ff', # 초록, #ffffff :하양, #000000: 검정
                    fill_color='#ffffff',
                    popup='Circle'
                    ).add_to(m) #지도에서 위치 크기 유지
m
```
